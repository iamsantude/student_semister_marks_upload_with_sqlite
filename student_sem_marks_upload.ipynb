# import the library 

```python
from tkinter import *
from tkinter.ttk import *
from tkinter import messagebox
import sqlite3 as sq
```

# Database Create and Table Create

```python
conn = sq.connect('student.db')
try:
    conn.execute('''CREATE TABLE MARKS
             (NAME   TEXT    NOT NULL,
             UNIVERSITY  TEXT     NOT NULL,
             COLLEGE   TEXT    NOT NULL,
             STREAM TEXT NOT NULL,
             SEM1 NUMBER(5,2) NOT NULL,
             SEM2 NUMBER(5,2) NOT NULL,
             SEM3 NUMBER(5,2) NOT NULL);''')
except:
    pass
try:
    conn.execute('''CREATE TABLE REGD
             (ID INTEGER PRIMARY KEY AUTOINCREMENT,
             NAME   TEXT    NOT NULL,
             AGE  INT     NOT NULL,
             COLLEGE   TEXT    NOT NULL,
             STREAM TEXT NOT NULL,
             GENDER TEXT NOT NULL,
             MAIL_ID TEXT NOT NULL,
             PASSW TEXT NOT NULL);''')
except:
    pass
```

# register window

```python
def register():
    rg = Tk()
    col1='dark salmon'
    col2='floral white'
    rg.resizable(False, False)
    rg.wm_attributes('-alpha', 0.95)
    rg.title("Registration")
    rg.configure(bg=col1)
    rg.geometry('500x500')
    style3 = Style() 
    style3.configure('W.TButton', font = 
                     ("Times New Roman Bold", 12, 'bold'), foreground = 'black',borderwidth = '3') 
    
    lbl1=Label(rg,text="Register",font=("Times New Roman Bold",30))
    lbl1.place(x=180,y=10)
    lbl1.configure(foreground=col2,background=col1)
    
    name=Label(rg,text="Name",font=("Times New Roman Bold",22))
    name.place(x=60,y=90)
    name.configure(foreground=col2,background=col1)
    nameent=Entry(rg,width=30) 
    nameent.place(x=210,y=103)

    age=Label(rg,text="Age",font=("Times New Roman Bold",22))
    age.place(x=60,y=130)
    age.configure(foreground=col2,background=col1)
    ageent=Entry(rg,width=30) 
    ageent.place(x=210,y=143)
    
    colg=Label(rg,text="College",font=("Times New Roman Bold",22))
    colg.place(x=60,y=170)
    colg.configure(foreground=col2,background=col1)
    colgcombo = Combobox(rg,width=27)
    colgcombo['values']= ("FIEM","UEM","HIT","BBIT")
    colgcombo.current(0)
    colgcombo.place(x=210,y=183)
    
    srt=Label(rg,text="Stream",font=("Times New Roman Bold",22))
    srt.place(x=60,y=210)
    srt.configure(foreground=col2,background=col1)
    srtcombo = Combobox(rg,width=27)
    srtcombo['values']= ("CSE","IT","MCA","ME")
    srtcombo.current(0)
    srtcombo.place(x=210,y=223)
    
    gen=Label(rg,text="Gender",font=("Times New Roman Bold",22))
    gen.place(x=60,y=250)
    gen.configure(foreground=col2,background=col1)
    gencombo = Combobox(rg,width=27)
    gencombo['values']= ("Male","Female")
    gencombo.current(0)
    gencombo.place(x=210,y=263)
    
    mail=Label(rg,text="Mail ID",font=("Times New Roman Bold",22))
    mail.place(x=60,y=290)
    mail.configure(foreground=col2,background=col1)
    mailent=Entry(rg,width=30) 
    mailent.place(x=210,y=303)
    
    pss=Label(rg,text="Password",font=("Times New Roman Bold",22))
    pss.place(x=60,y=330)
    pss.configure(foreground=col2,background=col1)
    pssent=Entry(rg,width=30) 
    pssent.place(x=210,y=343)
    
    def regfn():
        if '@' not in mailent.get():
            messagebox.showerror('Error', 'Wrong Mail ID')
        elif pssent.get().isalpha()==True or pssent.get().isdigit()==True:
            messagebox.showerror('Error', 'Password should be 8 character long\nPassword Sould Contain at least\nOne Upper Case\nOne Lower Case\nOne Digit')
        else:
            val=(nameent.get(),ageent.get(),colgcombo.get(),srtcombo.get(),gencombo.get(),mailent.get(),pssent.get())
            sql="INSERT INTO REGD (NAME,AGE,COLLEGE,STREAM,GENDER,MAIL_ID,PASSW) VALUES (?,?,?,?,?,?,?)"
            conn.execute(sql,val)
            conn.commit()
            messagebox.showinfo('Success','Detailed inserted for '+nameent.get())
            rg.destroy()
            login()

    regbtn=Button(rg,text="Register",command=regfn,style = 'W.TButton')
    regbtn.place(x=210,y=383)
    
    def clearfn():
        nameent.delete(0,END)
        ageent.delete(0,END)
        mailent.delete(0,END)
        pssent.delete(0,END)

    clrbtn=Button(rg,text="Clear",command=clearfn,style = 'W.TButton')
    clrbtn.place(x=310,y=383)
    
    rg.mainloop()
```

# Upload students deatels

```python
def details_upld():
    up = Tk()
    col1='blue'
    col2='floral white'
    up.resizable(False, False)
    up.wm_attributes('-alpha', 0.95)
    up.title("Upload Student Details")
    up.configure(bg=col1)
    up.geometry('500x500')
    style3 = Style() 
    style3.configure('W.TButton', font = 
                     ("Calibri", 14, 'bold'), foreground = 'black',borderwidth = '4')
    
    lbl1=Label(up,text="Upload",font=("Calibri",30))
    lbl1.place(x=180,y=10)
    lbl1.configure(foreground=col2,background=col1)
    
    name=Label(up,text="Name",font=("Calibri",22))
    name.place(x=60,y=90)
    name.configure(foreground=col2,background=col1)
    nameent=Entry(up,width=30) 
    nameent.place(x=230,y=103)
    
    univ=Label(up,text="University",font=("Calibri",22))
    univ.place(x=60,y=130)
    univ.configure(foreground=col2,background=col1)
    univent=Entry(up,width=30) 
    univent.place(x=230,y=143)
    
    clg=Label(up,text="College",font=("Calibri",22))
    clg.place(x=60,y=170)
    clg.configure(foreground=col2,background=col1)
    clgcombo = Combobox(up,width=27)
    clgcombo['values']= ("FIEM","UEM","HIT","BBIT")
    clgcombo.current(0)
    clgcombo.place(x=230,y=183)
    
    strm=Label(up,text="Stream",font=("Calibri",22))
    strm.place(x=60,y=210)
    strm.configure(foreground=col2,background=col1)
    strmcombo = Combobox(up,width=27)
    strmcombo['values']= ("CSE","IT","MCA","ME")
    strmcombo.current(0)
    strmcombo.place(x=230,y=223)
    
    sem1=Label(up,text="Sem1_Marks",font=("Calibri",22))
    sem1.place(x=60,y=250)
    sem1.configure(foreground=col2,background=col1)
    sem1ent=Entry(up,width=15) 
    sem1ent.place(x=230,y=263)
    
    sem2=Label(up,text="Sem2_Marks",font=("Calibri",22))
    sem2.place(x=60,y=300)
    sem2.configure(foreground=col2,background=col1)
    sem2ent=Entry(up,width=15) 
    sem2ent.place(x=230,y=313)
    
    sem3=Label(up,text="Sem3_Marks",font=("Calibri",22))
    sem3.place(x=60,y=350)
    sem3.configure(foreground=col2,background=col1)
    sem3ent=Entry(up,width=15) 
    sem3ent.place(x=230,y=363)
    
    def upld():
        try:
            val1=(nameent.get(),univent.get(),clgcombo.get(),strmcombo.get(),sem1ent.get(),sem2ent.get(),sem3ent.get())
            sql1="INSERT INTO MARKS (NAME,UNIVERSITY,COLLEGE,STREAM,SEM1,SEM2,SEM3) VALUES (?,?,?,?,?,?,?)"
            conn.execute(sql1,val1)           
            conn.commit()
            
            avg=(float(sem1ent.get())+float(sem2ent.get())+float(sem1ent.get()))/3
            messagebox.showinfo('Uploaded Details', 'Name: '+nameent.get()+'\nCollege: '+clgcombo.get()+'\nStream: '+strmcombo.get()+'\nAverage marks: '+str(avg))
           
            clfn()
        except:
            messagebox.showerror('Error','error in input!')
            
    uplbtn=Button(up,text="Upload",command=upld,style = 'W.TButton')
    uplbtn.place(x=220,y=423)
    
       
    def clfn():
        nameent.delete(0,END)
        univent.delete(0,END)
        sem1ent.delete(0,END)
        sem2ent.delete(0,END)
        sem3ent.delete(0,END)

    clbtn=Button(up,text="Clear",command=clfn,style = 'W.TButton')
    clbtn.place(x=350,y=423)
    
    up.mainloop()
```

# login window

```python
def login():
    windows = Tk()
    windows.resizable(False, False)
    windows.wm_attributes('-alpha', 0.9)
    windows.title("ID Validation")
    windows.configure(bg='cyan')
    windows.geometry('700x300')
    style3 = Style() 
    style3.configure('W.TButton', font = 
                     ("Times New Roman Bold", 12, 'bold'), foreground = 'black',borderwidth = '3') 

    lbl1=Label(windows,text="Login Window",font=("Times New Roman Bold",30))
    lbl1.place(x=250,y=10)
    lbl1.configure(foreground='black',background='cyan')

    name=Label(windows,text="ID",font=("Times New Roman Bold",20))
    name.place(x=200,y=100)
    name.configure(foreground='black',background='cyan')
    ent1=Entry(windows,width=30)
    ent1.place(x=270,y=107)

    pss=Label(windows,text="Pass",font=("Times New Roman Bold",20))
    pss.place(x=200,y=150)
    pss.configure(foreground='black',background='cyan')
    ent2=Entry(windows,width=30,show='*')    
    ent2.place(x=270,y=157)
    
    def loginchk():
        
        cursor=conn.execute("SELECT MAIL_ID,PASSW from REGD")
        mid=[]
        pas=[]
        for row in cursor:
            mid.append(row[0])
            pas.append(row[1])
        print("All mail id: ",mid)
        print("All password id: ",pas)
        try:
            ind=mid.index(ent1.get())
            authpass=pas[ind]
        except:
            ent1.delete(0,END)
            ent2.delete(0,END)
            messagebox.showerror('Error', 'Wrong ID')
    
        if ent1.get() in mid:
            if ent2.get()==authpass:
                messagebox.showinfo('Login Success', 'Welcome')
                windows.destroy()   
                details_upld()
        else:
            messagebox.showerror('Error', 'Wrong Credential')

    btn1=Button(windows,text="Login",command=loginchk,style = 'W.TButton')
    btn1.place(x=260,y=200)
    
    def refbtn():
        windows.destroy()
        login()

    btn2=Button(windows,text="Refresh",command=refbtn,style = 'W.TButton')
    btn2.place(x=370,y=200)

    reg=Label(windows,text="Not Registered?",font=("Times New Roman Bold",15))
    reg.place(x=250,y=250)
    reg.configure(foreground='black',background='cyan')
    
    def regbtn():
        windows.destroy()
        register()
    

    btn3=Button(windows,text="Register",command=regbtn,style = 'W.TButton')
    btn3.place(x=395,y=248)
    
    windows.mainloop()
login()
```
